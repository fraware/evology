## Does the current VI trades full wealth when stock trades at half value?

Let us assume that the valuation is 100. How does the trading signal varies with the price then?

```python
ValVI = 100
def PhiVI(p):
    return np.log2(ValVI / p)

x = np.linspace(0.1,250,1000)

# the function
VI = PhiVI(x)
VItanh = np.tanh(PhiVI(x)) + 0.5


fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.plot(x,VI, 'b', label='Phi_VI (TS only - Maarten model)')
plt.plot(x,VItanh, 'r', label='Tanh(Phi_VI) (Maarten model)')
plt.plot(50,1,'ro')
plt.legend(loc='upper right')
plt.axvline(x=100, color='black', linestyle=':')
plt.axhline(y=0, color='black', linestyle=':')
plt.axhline(y=1, color='black', linestyle=':')
plt.axhline(y=-1, color='black', linestyle=':')
plt.axhline(y=0.5, color='gray', linestyle=':')
plt.axhline(y=-0.5, color='gray', linestyle=':')
plt.axvline(x=200, color='gray', linestyle=':')
plt.axvline(x=50, color='gray', linestyle=':')
plt.show()

print('The trading signal alone satisfies the requirement. However, it is neither bounded, nor differentiable.')
print('However, the adjusted trading signal does not cross the point.')
print('In addition, we violate the [1,-1] interval.')

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.plot(x,VItanh, 'r', label='Tanh(Phi_VI) (Maarten model)')
plt.plot(50,1,'ro')
plt.legend(loc='upper right')
plt.axvline(x=100, color='black', linestyle=':')
plt.axhline(y=0, color='black', linestyle=':')
plt.axhline(y=1, color='black', linestyle=':')
plt.axhline(y=-1, color='black', linestyle=':')
plt.axhline(y=0.5, color='gray', linestyle=':')
plt.axhline(y=-0.5, color='gray', linestyle=':')
plt.axvline(x=200, color='gray', linestyle=':')
plt.axvline(x=50, color='gray', linestyle=':')
plt.show()
```

## RIght. Now, trend followers.
A symmetric assumption would be that TFs would use their full wealth if the price yesterday is at half of the day before.

This leads us to consider Pi as a candidate for TF aggression scale.

```python
ValVI = 100
def PhiVI(p):
    return np.log2(ValVI / p)

CST = 1 / 20

def NewVIfunc(p):
    return np.tanh( CST * (ValVI - p))

x = np.linspace(0.1,200,1000)

VI = PhiVI(x)
VItanh1 = np.tanh(PhiVI(x))
VItanhPi = np.tanh(np.pi * PhiVI(x))
NewVI = NewVIfunc(x)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.plot(x,VItanh1, 'r', label='Tanh(Phi_VI) (c=1)')
plt.plot(x, NewVI, 'y', label='newVI')
plt.plot(x,VItanhPi, 'g', label='Tanh(Phi_VI) (c=Pi)')
plt.plot(50,1,'ro')
plt.legend(loc='lower left')
plt.axvline(x=100, color='black', linestyle=':')
plt.axhline(y=0, color='black', linestyle=':')
plt.axhline(y=1, color='black', linestyle=':')
plt.axhline(y=-1, color='black', linestyle=':')
plt.axhline(y=0.5, color='gray', linestyle=':')
plt.axhline(y=-0.5, color='gray', linestyle=':')
plt.axvline(x=200, color='gray', linestyle=':')
plt.axvline(x=50, color='gray', linestyle=':')
plt.show()
```
