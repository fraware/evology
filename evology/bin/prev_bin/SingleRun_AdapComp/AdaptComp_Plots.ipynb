```python
#!/usr/bin/env python3
import pandas as pd
import random 
import matplotlib
import matplotlib.pyplot as plt 
data = pd.DataFrame()

rdsearch = pd.read_csv("/Users/aymericvie/Documents/GitHub/evology/evology/research/TransferStatus/SingleRun_AdapComp/data/rdsearch.csv")
profadap = pd.read_csv("/Users/aymericvie/Documents/GitHub/evology/evology/research/TransferStatus/SingleRun_AdapComp/data/profadapt.csv")
combined = pd.read_csv("/Users/aymericvie/Documents/GitHub/evology/evology/research/TransferStatus/SingleRun_AdapComp/data/combined.csv")
static = pd.read_csv("/Users/aymericvie/Documents/GitHub/evology/evology/research/TransferStatus/SingleRun_AdapComp/data/static.csv")
```

## Price

```python
''' Price '''
data['Price (Static)'] = static['Price']
data['Price (Random Search)'] = rdsearch['Price']
data['Price (Imitation)'] = profadap['Price']
data['Price (Combined)'] = combined['Price']

data['Gen'] = rdsearch['Gen']
data.plot(x="Gen", y = ["Price (Static)", "Price (Random Search)", "Price (Imitation)", "Price (Combined)"],
        kind="line", figsize=(15, 6), title='Comparison of price', 
        color = ['black', 'red','blue', 'green'],
        style = ['-', '-','-', '-'])
plt.xlabel('Time (days)')
plt.ylabel('Price')
plt.savefig('figures/adcomp_price.png', dpi = 300)
plt.show() 
```

## Volume

```python
''' Volume '''

data['Volume (Static)'] = static['Volume']
data['Volume (Random Search)'] = rdsearch['Volume']
data['Volume (Imitation)'] = profadap['Volume'] 
data['Volume (Combined)'] = combined['Volume'] 

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize = (10, 5), sharex = True, sharey = True)
fig.suptitle('Volume')
ax1.plot(data["Volume (Static)"], label = 'Volume (Static)')
ax2.plot(data["Volume (Random Search)"], label = 'Volume (Random Search)')
ax3.plot(data["Volume (Imitation)"], label = 'Volume (Imitation)')
ax4.plot(data["Volume (Combined)"], label = 'Volume (Combined)')
ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()
ax2.set_xlabel('Time (days)')
ax2.set_ylabel('Volume')
plt.savefig('figures/adcomp_vol.png', dpi = 300)
plt.show()
```

## Wealth shares

```python

data['WShare_NT (Random Search)'] = rdsearch['WShare_NT']
data['WShare_VI (Random Search)'] = rdsearch['WShare_VI']
data['WShare_TF (Random Search)'] = rdsearch['WShare_TF']

data['WShare_NT (Imitation)'] = profadap['WShare_NT']
data['WShare_VI (Imitation)'] = profadap['WShare_VI']
data['WShare_TF (Imitation)'] = profadap['WShare_TF']

data['WShare_NT (Combined)'] = combined['WShare_NT']
data['WShare_VI (Combined)'] = combined['WShare_VI']
data['WShare_TF (Combined)'] = combined['WShare_TF']

data['WShare_NT (Static)'] = static['WShare_NT']
data['WShare_VI (Static)'] = static['WShare_VI']
data['WShare_TF (Static)'] = static['WShare_TF']

fig, (ax0, ax1, ax2, ax3) = plt.subplots(4, 1, figsize = (15, 7), sharex = True, sharey = True)
fig.suptitle('Strategy wealth share and population size')

ax0.plot(data['WShare_NT (Static)'], label = 'WShare_NT (Static)')
ax0.plot(data['WShare_VI (Static)'], label = 'WShare_VI (Static)')
ax0.plot(data['WShare_TF (Static)'], label = 'WShare_TF (Static)')

ax1.plot(data['WShare_NT (Random Search)'], label = 'WShare_NT (Random Search)')
ax1.plot(data['WShare_VI (Random Search)'], label = 'WShare_VI (Random Search)')
ax1.plot(data['WShare_TF (Random Search)'], label = 'WShare_TF (Random Search)')

ax2.plot(data['WShare_NT (Imitation)'], label = 'WShare_NT (Imitation)')
ax2.plot(data['WShare_VI (Imitation)'], label = 'WShare_VI (Imitation)')
ax2.plot(data['WShare_TF (Imitation)'], label = 'WShare_TF (Imitation)')

ax3.plot(data['WShare_NT (Combined)'], label = 'WShare_NT (Combined)')
ax3.plot(data['WShare_VI (Combined)'], label = 'WShare_VI (Combined)')
ax3.plot(data['WShare_TF (Combined)'], label = 'WShare_TF (Combined)')

ax1.legend(loc = 1)
ax2.legend(loc = 1)
ax3.legend(loc = 1)
ax2.set_xlabel('Time (days)')
ax2.set_ylabel('Wealth shares (%)')
plt.savefig('figures/adcomp_wshare.png', dpi = 300)
plt.show()
```
