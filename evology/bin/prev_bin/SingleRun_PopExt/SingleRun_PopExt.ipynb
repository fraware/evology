```python

import pandas as pd
import random 
import matplotlib
import matplotlib.pyplot as plt 
data = pd.DataFrame()

small = pd.read_csv("/Users/aymericvie/Documents/GitHub/evology/evology/research/TransferStatus/SingleRun_PopExt/data/small_15k.csv")
med = pd.read_csv("/Users/aymericvie/Documents/GitHub/evology/evology/research/TransferStatus/SingleRun_PopExt/data/medium_15k.csv")
big = pd.read_csv("/Users/aymericvie/Documents/GitHub/evology/evology/research/TransferStatus/SingleRun_PopExt/data/big_15k.csv")
```

## Price

```python
''' Price '''
data['Price (Small Pop)'] = small['Price']
data['Price (Medium Pop)'] = med['Price']
data['Price (Large Pop)'] = big['Price']
data['Gen'] = small['Gen']
data.plot(x="Gen", y = ["Price (Small Pop)", "Price (Medium Pop)", "Price (Large Pop)"],
        kind="line", figsize=(15, 6), title='Comparison of price', 
        color = ['red','blue', 'green'],
        style = ['-','-', '-'])
plt.xlabel('Time (days)')
plt.ylabel('Price')
plt.savefig('figures/ext_price.png', dpi = 300)
plt.show() 
```

## Volume

```python
''' Volume '''

data['Volume (Small Pop)'] = small['Volume']
data['Volume (Medium Pop)'] = med['Volume'] / 10
data['Volume (Large Pop)'] = big['Volume'] / (10/3)

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize = (10, 5), sharex = True, sharey = True)
fig.suptitle('Volume (adjusted for population sizes)')
ax1.plot(data["Volume (Small Pop)"], label = 'Volume (Small Pop)')
ax2.plot(data["Volume (Medium Pop)"], label = 'Volume (Medium Pop)')
ax3.plot(data["Volume (Large Pop)"], label = 'Volume (Large Pop)')
ax1.legend()
ax2.legend()
ax3.legend()
ax2.set_xlabel('Time (days)')
ax2.set_ylabel('Volume')
plt.savefig('figures/ext_vol.png', dpi = 300)
plt.show()
```
